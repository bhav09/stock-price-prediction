Configure local MLflow tracking store. Manual logging (autolog disabled) for parameters, metrics, and artifacts.


Default: logs to `mlruns/`. Set `MLFLOW_TRACKING_URI` to change location.

```python
import os, mlflow, tempfile, json

# Configure MLflow tracking store and experiment name
os.environ.setdefault("MLFLOW_TRACKING_URI", "mlruns")
mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI"))
mlflow.set_experiment(os.getenv("MLFLOW_EXPERIMENT", "stock_pred"))
mlflow.autolog(disable=True)
mlflow
```

```python
# Display MLflow configuration
print(f"MLflow Tracking URI: {os.getenv('MLFLOW_TRACKING_URI')}")
print(f"MLflow Experiment: {os.getenv('MLFLOW_EXPERIMENT', 'stock_pred')}")
print(f"\nMLflow module: {mlflow}")
print(f"MLflow version: {mlflow.__version__}")
```

**Test run:** Log sample params/metrics/artifacts. View with `mlflow ui`.

```python

# Create a test run to verify MLflow logging functionality
with mlflow.start_run(run_name="smoke_test"):
    mlflow.log_param("example_param", 123)
    mlflow.log_metric("example_metric", 0.42)
    
    # Log artifacts (files) to MLflow for tracking model outputs or other files
    with tempfile.TemporaryDirectory() as td:
        open(os.path.join(td, "hello.txt"), "w").write("hi mlflow")
        mlflow.log_artifacts(td, "artifacts")
print("done")
```
